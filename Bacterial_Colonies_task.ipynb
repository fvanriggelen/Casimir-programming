```python
# Import modules
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from PIL import Image #to import and adapt images (Python Image Library)
from scipy.misc import imread as sc_imread
import scipy.ndimage

%matplotlib inline
```

```python
"""Import the image and display it"""

fig = plt.figure(figsize = (12,7))
ax = fig.add_subplot(111)

#provide the path of the image that you want to open and open the image
image_path = "../img/results_L3.jpg"
im = Image.open(image_path);
ax.imshow(im)
```

```python
"""Convert image to BW """
#There are several ways to do this, here we use the convert method in image

fig = plt.figure(figsize = (12,7))
ax = fig.add_subplot(111)
im_gray = im.convert('L')
plt.imshow(im_gray, cmap=cm.Greys_r)

```
